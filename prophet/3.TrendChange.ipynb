```python
import pandas as pd
import numpy as np
from fbprophet import Prophet

from quilt.data.examples import prophet as pdata

%matplotlib inline
```

* Changepoints are "inflection" points in the data
* Prophet will auto-detect, but we can also customize (using `n_changepoints` or, preferred, `change_prior_scale`)
* The sparsity of `change_prior_scale` dictates that few(er) of the guessed changepoints will actually be used

```python
df = pdata.peyton_manning()
df['y'] = np.log(df['y'])
m = Prophet(changepoint_prior_scale=0.01)
m.fit(df)
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
m.plot(forecast);
```

```python
df = pdata.peyton_manning()
df['y'] = np.log(df['y'])
m = Prophet(changepoint_prior_scale=0.5) # override 0.05 default => more flexible trend
m.fit(df)
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
m.plot(forecast);
```

```python
#makes changepoints less frequent
m = Prophet(changepoint_prior_scale=0.001)
forecast = m.fit(df).predict(future)
m.plot(forecast);
```

```python
m = Prophet(changepoints=['2014-01-01'])
forecast = m.fit(df).predict(future)
m.plot(forecast);
```

```python
m = Prophet(weekly_seasonality=False)
m.add_seasonality(name='monthly', period=30.5, fourier_order=5)
forecast = m.fit(df).predict(future)
m.plot_components(forecast);
```
